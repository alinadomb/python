```python
!pip install mysql-connector-python

import mysql.connector as connection

try:
    mydb = connection.connect(host="localhost", database = 'employees',user="root", passwd="Admin#123",use_pure=True)
    query = '''WITH CTEsal(emp_no, sal_month, dept_no, sum_sal) AS
    (SELECT s.emp_no, (s.salary/12) AS sal_month, de.dept_no,
    SUM(s.salary/12) OVER (PARTITION BY de.dept_no ORDER BY s.salary/12 ASC 
    RANGE BETWEEN UNBOUNDED  PRECEDING AND CURRENT ROW) AS sum_sal
    FROM salaries AS s
    INNER JOIN dept_emp AS de USING(emp_no)
    WHERE CURDATE() BETWEEN s.from_date AND s.to_date
    AND CURDATE() BETWEEN de.from_date AND de.to_date
    ORDER BY de.dept_no ASC, sum_sal ASC)
  

    SELECT e.emp_no, e.first_name, e.last_name, c.dept_no, c.sum_sal
    FROM employees AS e
    INNER JOIN CTEsal AS c USING(emp_no)
    WHERE c.sum_sal <= 500000;'''
    result_dataFrame = pd.read_sql(query,mydb)
    mydb.close() #close the connection
except Exception as e:
    mydb.close()
    print(str(e))

result_dataFrame
```
